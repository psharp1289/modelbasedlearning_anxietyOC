```python
# Calculate Gamma shape and rate from mode and sd.
def gammaShRaFromModeSD(mode, sd):
    rate = (mode + np.sqrt( mode**2 + 4 * sd**2 ) ) / ( 2 * sd**2 )
    shape = 1 + mode * rate
    return(shape, rate)

ySD=np.std(choice_data)
yMean=np.mean(choice_data)
print(yMean)
gammas = gammaShRaFromModeSD(ySD/2, 2*ySD)
bl_sd=1/((ySD*5)**2)
print(bl_sd)
gammas=list(gammas)
noise=[ySD/100,ySD*10]
print(noise)
print(gammas)
```

#### For model below, i=condition, j=subject, g=group (0: low psychopathology, 1: high psychopathology)

$data_{i,j} \sim Bernoulli(\theta_{i,j})$

$\theta_{i,j}= logistic(baseline+subject_{j}+condition_{i}+psychopathology_{g}+interaction_{i,g})$

$subject{j} \sim \mathcal{N}(0,\sigma_{subject})$

$condition{i} \sim \mathcal{N}(0,\sigma_{condition})$

$psychopathology_{g} \sim \mathcal{N}(0,\sigma_{psychopathology})$

$interaction_{i,g} \sim \mathcal{N}(0,\sigma_{interaction})$

$\sigma_{all} U(0,5)$

Convert all to sum-to-zero

# Anxious Arousal model

```python
from pystan import StanModel


model_input='''
data {

    int<lower=1> NxConditions;
    int<lower=1> NxSubjects;
    int Choice[NxSubjects, NxConditions];
    int<lower=0,upper=1> group[NxSubjects];
}

parameters {   
  // Grand mean
  real bl;
  vector[NxSubjects] a0;
  real<lower=0.001,upper=0.999> sigma_subject;
  
  //Conditions
  vector[NxConditions] aW; 
  matrix[NxConditions,2] aInteraction;
  vector[2] aGroup;
  
}

transformed parameters {

vector<lower=0,upper=1>[NxSubjects] theta_neutral;
vector<lower=0,upper=1>[NxSubjects] theta_cp;
vector<lower=0,upper=1>[NxSubjects] theta_ct;
vector<lower=0,upper=1>[NxSubjects] theta_rp;
vector<lower=0,upper=1>[NxSubjects] theta_rt;



for (k in 1:NxSubjects){
    if (group[k]==0){
        theta_neutral[k] = inv_logit(bl+a0[k]+aW[1]+aGroup[1]+aInteraction[1,1]);
        theta_cp[k] = inv_logit(bl+a0[k]+aW[2]+aGroup[1]+aInteraction[2,1]);
        theta_ct[k] = inv_logit(bl+a0[k]+aW[3]+aGroup[1]+aInteraction[3,1]);
        theta_rp[k] = inv_logit(bl+a0[k]+aW[4]+aGroup[1]+aInteraction[4,1]);
        theta_rt[k] = inv_logit(bl+a0[k]+aW[5]+aGroup[1]+aInteraction[5,1]);}
    else if (group[k]==1){
        theta_neutral[k] = inv_logit(bl+a0[k]+aW[1]+aGroup[2]+aInteraction[1,2]);
        theta_cp[k] = inv_logit(bl+a0[k]+aW[2]+aGroup[2]+aInteraction[2,2]);
        theta_ct[k] = inv_logit(bl+a0[k]+aW[3]+aGroup[2]+aInteraction[3,2]);
        theta_rp[k] = inv_logit(bl+a0[k]+aW[4]+aGroup[2]+aInteraction[4,2]);
        theta_rt[k] = inv_logit(bl+a0[k]+aW[5]+aGroup[2]+aInteraction[5,2]);
    }
    }
}
model {    
    sigma_subject~uniform(0,1);
    bl ~ normal(0.6,0.10);
    a0 ~ normal(0,sigma_subject);
    aW ~ normal(0,1);
    aGroup ~ normal(0,1);
    aInteraction[:,1] ~ normal(0,1);
    aInteraction[:,2] ~ normal(0,1);
    
    for (k in 1:NxSubjects){
        // Neutral condition choices
        Choice[k,1] ~ bernoulli(theta_neutral[k]);

        // Positive-common choices
        Choice[k,2] ~ bernoulli(theta_cp[k]);

        // Threat-common choices
        Choice[k,3] ~ bernoulli(theta_ct[k]); 

        // Positive-rare choices
        Choice[k,4] ~ bernoulli(theta_rp[k]);

        // Threat-rare choices
        Choice[k,5] ~ bernoulli(theta_rt[k]);
     
          }    
      
    }
    
'''
data_input = {
              'Choice':choice_data,#choice data
              'NxConditions':5,
              'NxSubjects':num_subjects,
              'group':clin_cutoff_MASQ
    
}

controls={'adapt_delta':0.9,
         'max_treedepth':18}

model_fit_Anova = StanModel(model_code=model_input)
fit_model_fullAnova_nocontrol = model_fit_Anova.sampling(data=data_input,iter=2000,control=controls)
```

```python
## Full ANOVA model with MASQ as individual difference measure (median split)

import seaborn as sns
from bayesian_bootstrap.bootstrap import mean, highest_density_interval
import pandas as pd
import matplotlib.pyplot as plt
import numpy.ma as ma
import numpy as np

fit_df_2 = fit_model_fullAnova_nocontrol.to_dataframe()

NxSubjectLvl=174
NxBetweenLvl=2
NxWithinLvl=5
NxSamples=len(fit_df_2['bl'])
print('number_samples {}'.format(NxSamples))
W=np.zeros((NxSubjectLvl, NxBetweenLvl, NxWithinLvl,NxSamples))
# Initialize the array with zeros
mSxBxW = np.zeros((NxSubjectLvl, NxBetweenLvl, NxWithinLvl, NxSamples))

# Fill the arrray
for k in range(NxSubjectLvl):
    for i in range(NxWithinLvl):
        for j in range(NxBetweenLvl):
            exec('mu = fit_df_2["bl"]+ fit_df_2["aGroup[{}]"] + fit_df_2["aW[{}]"] + fit_df_2["aInteraction[{},{}]"] + fit_df_2["a0[{}]"]'.format(j+1,i+1,i+1,j+1,k+1))
            mSxBxW[k,j,i,:] = mu

# Convert to masked array that masks value '0'.
mSxBxW_ma = ma.masked_equal(mSxBxW, 0)
print('before/after mask')
print(mSxBxW.shape)
print(mSxBxW_ma.shape)
print('')

# Mean for subject S across levels of W, within the level of B
mS = ma.mean(mSxBxW_ma, axis=(2))
print('ms shape')
print(mS.data.shape)

# Mean for treatment combination BxW, across subjects S
mBxW = mSxBxW_ma.mean(axis=(0))
print('mBxW shape')
print(mBxW.data.shape)

# Mean for level B, across W and S
# Keeping the dimension of axis 1 in order to have the broadcasting work correctly when calculating bBxW.
mB = ma.mean(mBxW, axis=(1), keepdims=True)
print('mb shape')
print(mB.shape)
# Mean for level W, across B and S
# Keeping the dimension of axis 0 in order to have the broadcasting work correctly when calculating bBxW.
print('mw shape')
mW = ma.mean(mBxW, axis=(0), keepdims=True)
print(mW.data.shape)
# Equation 20.3
m = ma.mean(mBxW, axis=(0,1))
b0 = m

print('Mean baseline yield: {}'.format(b0.mean()))
print(b0.shape)
# Equation 20.4
# Suppress the dimension with size one for mB in order to have the broadcasting work properly.
bB = mB.squeeze() - m

# Equation 20.5
# Suppress the dimension with size one for mW in order to have the broadcasting work properly.
bW = mW.squeeze() - m

# CHECK SUM-TO-ZERO WORKED
# print('MEANS MAIN EFFECTS:')
# print(np.mean(bW[0,:]))
# print(np.mean(bW[1,:]))
# print(np.mean(bW[2,:]))
# print('')
# print('SUM OF MAIN EFFECTS:')
# print(np.mean(bW[0,:])+np.mean(bW[1,:])+np.mean(bW[2,:]))
# print('')

# Equation 20.6
bBxW = mBxW - mB - mW + m
print(bBxW.shape)

# Equation 20.7
bS = mS - mB.squeeze()

#Extract data

neutral = bW[0,:]
cp = bW[1,:]
ct = bW[2,:]
rp = bW[3,:]
rt = bW[4,:]

low_anx=bB[0,:]
high_anx=bB[1,:]

diff_rt = neutral-rt

#interactions
n_low = bBxW[0,0,:]
cp_low = bBxW[0,1,:]
ct_low = bBxW[0,2,:]
rp_low = bBxW[0,3,:]
rt_low = bBxW[0,4,:]

n_high = bBxW[1,0,:]
cp_high = bBxW[1,1,:]
ct_high = bBxW[1,2,:]
rp_high = bBxW[1,3,:]
rt_high = bBxW[1,4,:]

print('mean n high : {}'.format(np.mean(n_high)))
print('mean cp high : {}'.format(np.mean(cp_high)))
print('mean ct high : {}'.format(np.mean(ct_high)))
print('mean rp high : {}'.format(np.mean(rp_high)))
print('mean rt high : {}'.format(np.mean(rt_high)))


print('mean n low : {}'.format(np.mean(n_low)))
print('mean cp low : {}'.format(np.mean(cp_low)))
print('mean ct low : {}'.format(np.mean(ct_low)))
print('mean rp low : {}'.format(np.mean(rp_low)))
print('mean rt low : {}'.format(np.mean(rt_low)))


interaction_low_ocd=((ct_low-rt_low)-(cp_low-rp_low))
interaction_high_ocd=((ct_high-rt_high)-(cp_high-rp_high))
#difference between neutral and rare threat
diff_low_high=interaction_high_ocd-interaction_low_ocd
common_t_high=ct_high
common_t_low=ct_low
common_emotion_interaction=common_t_high-common_t_low
rare_t_high=rt_high
rare_t_low=rt_low
rare_emotion_interaction=rare_t_high-rare_t_low
common_positive_int=cp_high-cp_low
rare_positive_int=rp_high-rp_low
diff_anx=low_anx-high_anx
z=0
for i in diff_anx:
    if i<0:
        z+=1
print('percentage of parameters above ROPE = {}'.format((8000-z)/8000.0))
print('')
rope=0.01

#Compute HDIs
ml_ct, mr_ct = highest_density_interval(common_emotion_interaction)
ml_cp, mr_cp = highest_density_interval(rare_emotion_interaction)
ml, mr = highest_density_interval(diff_low_high)
hml, hmr = highest_density_interval(diff_anx)
ft,fr=highest_density_interval(common_positive_int)
gt,gr=highest_density_interval(rare_positive_int)
print(mr)
print(hml)
print('')
print(np.mean(neutral))
print(np.mean(cp))
print(np.mean(ct))
print(np.mean(rt))
print(np.mean(rp))
print(np.mean(low_anx))

sns.set(style="white",
        palette="muted",
        color_codes=True)

f, axs = plt.subplots(6, 1, figsize=(8, 20),sharex=False)



ax1=sns.distplot(diff_low_high, hist=False,kde_kws={"shade": True},color="b", ax=axs[0])
ax1.set(title='interaction',xlabel='')

axs[0].plot([ml, mr],[0,0],linewidth=8.0,label='95% HDI',marker='o',color='y')
axs[0].plot([-rope, rope],[0,0],linewidth=8.0,label='ROPE',marker='o',color='black')

ax1=sns.distplot(diff_anx, hist=False,kde_kws={"shade": True},color="b", ax=axs[1])
ax1.set(title='',xlabel='')
x,y = ax1.get_lines()[0].get_data()
y=list(y)
x=list(x)
mode= max(y)
index_mode=y.index(max(y))
print('mode of dist = {}'.format(mode))
print('index of mode: ={}'.format(index_mode))
print('mode of x value: = {}'.format(x[index_mode]))

axs[1].plot([hml,hmr],[0,0],linewidth=8.0,label='95% HDI',marker='o',color='y')
axs[1].plot([-rope, rope],[0,0],linewidth=8.0,label='ROPE',marker='o',color='black')
print(hml, hmr)
ax0=sns.distplot(common_emotion_interaction, hist=False,kde_kws={"shade": True},color="r", ax=axs[2])
ax0.set(title='effect of common threat across low and high anxious arousal',xlabel='')

axs[2].plot([ml_ct, mr_ct],[0,0],linewidth=8.0,label='95% HDI',marker='o',color='y')
axs[2].plot([-rope, rope],[0,0],linewidth=8.0,label='ROPE',marker='o',color='black')

ax1=sns.distplot(rare_emotion_interaction, hist=False,kde_kws={"shade": True},color="b", ax=axs[3])
ax1.set(title='effect of rare threat across low and high anxious arousal',xlabel='')

axs[3].plot([ml_cp, mr_cp],[0,0],linewidth=8.0,label='95% HDI',marker='o',color='y')
axs[3].plot([-rope, rope],[0,0],linewidth=8.0,label='ROPE',marker='o',color='black')


ax0=sns.distplot(common_positive_int, hist=False,kde_kws={"shade": True},color="r", ax=axs[4])
ax0.set(title='effect of common positive across low and high anxious arousal',xlabel='')

axs[4].plot([ft, fr],[0,0],linewidth=8.0,label='95% HDI',marker='o',color='y')
axs[4].plot([-rope, rope],[0,0],linewidth=8.0,label='ROPE',marker='o',color='black')

ax1=sns.distplot(rare_positive_int, hist=False,kde_kws={"shade": True},color="b", ax=axs[5])
ax1.set(title='effect of rare positive across low and high anxious arousal',xlabel='')

axs[5].plot([gt, gr],[0,0],linewidth=8.0,label='95% HDI',marker='o',color='y')
axs[5].plot([-rope, rope],[0,0],linewidth=8.0,label='ROPE',marker='o',color='black')

plt.savefig('high_res_anxious_arousal_interactions.png', dpi=300)

plt.tight_layout()
plt.show()
```

```python
from pystan import StanModel


model_input='''
data {

    int<lower=1> NxConditions;
    int<lower=1> NxSubjects;
    int Choice[NxSubjects, NxConditions];
    int<lower=0,upper=1> group[NxSubjects];
}

parameters {   
  // Grand mean
  real bl;
  vector[NxSubjects] a0;

  
  //Conditions
  vector[NxConditions] aW; 
  matrix[NxConditions,2] aInteraction;
  vector[2] aGroup;
  
}

transformed parameters {

vector<lower=0,upper=1>[NxSubjects] theta_neutral;
vector<lower=0,upper=1>[NxSubjects] theta_cp;
vector<lower=0,upper=1>[NxSubjects] theta_ct;
vector<lower=0,upper=1>[NxSubjects] theta_rp;
vector<lower=0,upper=1>[NxSubjects] theta_rt;



for (k in 1:NxSubjects){
    if (group[k]==0){
        theta_neutral[k] = inv_logit(bl+a0[k]+aW[1]+aGroup[1]+aInteraction[1,1]);
        theta_cp[k] = inv_logit(bl+a0[k]+aW[2]+aGroup[1]+aInteraction[2,1]);
        theta_ct[k] = inv_logit(bl+a0[k]+aW[3]+aGroup[1]+aInteraction[3,1]);
        theta_rp[k] = inv_logit(bl+a0[k]+aW[4]+aGroup[1]+aInteraction[4,1]);
        theta_rt[k] = inv_logit(bl+a0[k]+aW[5]+aGroup[1]+aInteraction[5,1]);}
    else if (group[k]==1){
        theta_neutral[k] = inv_logit(bl+a0[k]+aW[1]+aGroup[2]+aInteraction[1,2]);
        theta_cp[k] = inv_logit(bl+a0[k]+aW[2]+aGroup[2]+aInteraction[2,2]);
        theta_ct[k] = inv_logit(bl+a0[k]+aW[3]+aGroup[2]+aInteraction[3,2]);
        theta_rp[k] = inv_logit(bl+a0[k]+aW[4]+aGroup[2]+aInteraction[4,2]);
        theta_rt[k] = inv_logit(bl+a0[k]+aW[5]+aGroup[2]+aInteraction[5,2]);
    }
    }
}
model {    

    bl ~ normal(0,0.25);
    a0 ~ normal(0,1);
    aW ~ normal(0,1);
    aGroup ~ normal(0,1);
    aInteraction[:,1] ~ normal(0,1);
    aInteraction[:,2] ~ normal(0,1);
    
    for (k in 1:NxSubjects){
        // Neutral condition choices
        Choice[k,1] ~ bernoulli(theta_neutral[k]);

        // Positive-common choices
        Choice[k,2] ~ bernoulli(theta_cp[k]);

        // Threat-common choices
        Choice[k,3] ~ bernoulli(theta_ct[k]); 

        // Positive-rare choices
        Choice[k,4] ~ bernoulli(theta_rp[k]);

        // Threat-rare choices
        Choice[k,5] ~ bernoulli(theta_rt[k]);
     
          }    
      
    }
    
'''
data_input = {
              'Choice':choice_data,#choice data
              'NxConditions':5,
              'NxSubjects':num_subjects,
              'group':median_split_PSWQ
}

controls={'adapt_delta':0.9,
         'max_treedepth':18}

model_fit_Anova = StanModel(model_code=model_input)
fit_model_fullAnova_nocontrol_worry = model_fit_Anova.sampling(data=data_input,iter=4000,control=controls)
```
