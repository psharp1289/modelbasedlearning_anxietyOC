# Relationship with True Variable when tradeoff variables are anticorrelated

```python
import numpy as np
from random import randrange
import seaborn as sns
import matplotlib.pyplot as plt
from numpy.random import normal as normal
from scipy.stats import pearsonr as rs
from random import uniform as u
import random


print('tradeoff when anticorrelated due to Y1*Y2=X')

#Y1 and Y2 are anticorrelated by virtue of Y1*Y2 = X
correl_xy1=[]
correl_xy2=[]
correl_y1y2=[]


for i in range(100):
    y1=[i for i in np.arange(1,500)]
    rev_order=sorted(y1,reverse=True)
    order1=[]
    for i in range(len(rev_order)):
        if i%10==0:
            order1.append(rev_order[i])
        else:
            order1.append(randrange(1,500))

    y2=[]
    x=[]
    for g in range(len(y1)):
        y2.append(y1[order1[g]-1])
        x.append((y1[g]*y2[g]))
    
 
    r,p=rs(x,y1)
    correl_xy1.append(r)
    r3,p3=rs(y2,x)
    correl_xy2.append(r3)
    r2,p2=rs(y1,y2)
    correl_y1y2.append(r2)


    
f, axs = plt.subplots(3, 1, figsize=(5, 5),sharex=True)

ax0=sns.distplot(correl_xy1, hist=False,kde_kws={"shade": True},color="k", ax=axs[0])
ax0.set(title='Correlation X and Y1',xlabel='')


ax1=sns.distplot(correl_xy2, hist=False,kde_kws={"shade": True},color="g", ax=axs[1])
ax1.set(title='Correlation X and Y2',xlabel='')

ax1=sns.distplot(correl_y1y2, hist=False,kde_kws={"shade": True},color="m", ax=axs[2])
ax1.set(title='Correlation Y1 and Y2',xlabel='')

plt.tight_layout()
plt.show()


```

# Relationship with True Variable when tradeoff variables are positively correlated

```python
print('tradeoff when pos correlated due to Y1/Y2=X')


reg_order=np.arange(1,500)
y1=np.arange(1,500)
correl_xy1=[]
correl_xy2=[]
correl_y1y2=[]
for i in range(500):
    order1=[]
    for i in range(len(rev_order)):
        if i%5==0:
            order1.append(reg_order[i])
        else:
            order1.append(randrange(1,500))

    y2=[]
    x=[]
    for g in range(len(y1)):
        y2.append(y1[order1[g]-1]/0.9)
        noise=np.random.normal(0,)
        x.append((y1[g]/y2[g])+noise)
    

    r,p=rs(x,y1)
    correl_xy1.append(r)
    r3,p3=rs(y2,x)
    correl_xy2.append(r3)
    r2,p2=rs(y1,y2)
    correl_y1y2.append(r2)

f, axs = plt.subplots(3, 1, figsize=(8, 5),sharex=True)

ax0=sns.distplot(correl_xy1, hist=False,kde_kws={"shade": True},color="k", ax=axs[0])
ax0.set(title='Correlation X and Y1',xlabel='')


ax1=sns.distplot(correl_xy2, hist=False,kde_kws={"shade": True},color="g", ax=axs[1])
ax1.set(title='Correlation X and Y2',xlabel='')

ax1=sns.distplot(correl_y1y2, hist=False,kde_kws={"shade": True},color="m", ax=axs[2])
ax1.set(title='Correlation Y1 and Y2',xlabel='')

plt.tight_layout()
plt.show()
```
